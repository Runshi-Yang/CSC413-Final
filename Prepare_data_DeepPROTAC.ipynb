```python
import os
from pathlib import  Path
import glob
import pandas as pd
import pickle
import numpy as np
from pymol import cmd
from tqdm import tqdm
from prepare_data import GraphData
```

```python
PROTAC_complex_path = 'small_dataset/protacs' # the complex file path, you should change.
PROTAC_data_path = 'small_dataset/protacs.csv' # the csv file path, including the label information, you should change.
```

```python
name_list = os.listdir(PROTAC_complex_path)
name_list = [x.split('.')[0] for x in name_list]
```

```python
name_list[:5]
```

```python
Path('small_dataset/ligase_ligand').mkdir(exist_ok=True)
Path('small_dataset/ligase_pocket_5').mkdir(exist_ok=True)
Path('small_dataset/target_ligand').mkdir(exist_ok=True)
Path('small_dataset/target_pocket_5').mkdir(exist_ok=True)
```

```python
with open('small_dataset/name.pkl','wb') as f:
    pickle.dump(name_list, f)
```

```python
for i in name_list:
    cmd.load(glob.glob('protacs/'+i+"/*igase.pdb")[0])
    cmd.remove('h.')
    cmd.select("small_dataset/ligase_ligand","resn res")
    cmd.save("small_dataset/ligase_ligand/"+i+".mol2","ligase_ligand")
    cmd.select("small_dataset/ligase_pocket_5","byres resn res around 5 and Chain A")
    cmd.save("small_dataset/ligase_pocket_5/"+i+".mol2","ligase_pocket_5")
    cmd.delete("all")

    cmd.load(glob.glob('protacs/'+i+"/*rotein.pdb")[0])
    cmd.remove('h.')
    cmd.select("small_dataset/target_ligand","resn res")
    cmd.save("small_dataset/target_ligand/"+i+".mol2","target_ligand")
    cmd.select("small_dataset/target_pocket_5","byres resn res around 5 and Chain B")
    cmd.save("small_dataset/target_pocket_5/"+i+".mol2","target_pocket_5")
    cmd.delete("all")
    
```
